```python
# 1. Покупатели, совершающие покупки еженедельно, предпочитают доступные товары по цене (ниже среднего), 
# чем те, кто покупает реже. 

from scipy.stats import ttest_ind
import plotly.graph_objects as go

def interpret(p_value):
    if p_value < 0.05:
        print("Есть статистически значимая зависимость между ними")
    else:
        print("Нет статистически значимой зависимости между ними")

data = df.copy()

# Разделим данные на две группы по частоте покупок
weekly_buyers = data[data['Frequency of Purchases'] == 'Weekly']
non_weekly_buyers = data[data['Frequency of Purchases'] != 'Weekly']

# Вычислиv среднюю стоимость покупки в обеих группах
avg_price_weekly = weekly_buyers['Purchase Amount (USD)'].mean()
avg_price_non_weekly = non_weekly_buyers['Purchase Amount (USD)'].mean()

print(f"Средняя стоимость покупки (еженедельные покупатели): {avg_price_weekly:.2f} USD")
print(f"Средняя стоимость покупки (остальные покупатели): {avg_price_non_weekly:.2f} USD")

# Проведем t-тест для проверки различий между средними
t_stat, p_value = ttest_ind(
    weekly_buyers['Purchase Amount (USD)'],
    non_weekly_buyers['Purchase Amount (USD)'],
    equal_var=False  # Предполагаем разные дисперсии
)

print(f"T-статистика: {t_stat:.2f}")
print(f"P-значение: {p_value:.4f}")

# Интерпретация результата
interpret(p_value)

# Визуализируем данные
categories = ['Еженедельные покупатели', 'Остальные покупатели']
values = [avg_price_weekly, avg_price_non_weekly]

fig = go.Figure(data=[
    go.Bar(name='Средняя стоимость покупки', x=categories, y=values)
])

fig.update_layout(
    title="Сравнение средней стоимости покупок между еженедельными покупателями и остальными",
    xaxis_title="Группы покупателей",
    yaxis_title="Средняя стоимость покупки (USD)",
    template="plotly"
)

fig.show()
```

```python
# Клиенты с активной подпиской совершают покупки чаще, чем те, у кого подписки нет 

# Преобразуем частоты покупок в числовой формат
frequency_map = {
    'Weekly': 52,
    'Fortnightly': 26,
    'Bi-Weekly': 26,
    'Monthly': 12,
    'Quarterly': 4,
    'Every 3 Months': 4,
    'Annually': 1
}
data['Frequency Numeric'] = data['Frequency of Purchases'].map(frequency_map)

# Разделим данные на подписчиков и неподписчиков и посчитаем количество покупателей для каждой частоты покупок
subscribers = data[data['Subscription Status'] == 'Yes']
non_subscribers = data[data['Subscription Status'] == 'No']

subscriber_counts = subscribers['Frequency of Purchases'].value_counts()
non_subscriber_counts = non_subscribers['Frequency of Purchases'].value_counts()

# Объединяем данные для визуализации
freq_options = list(frequency_map.keys())
subscriber_counts = subscriber_counts.reindex(freq_options, fill_value=0)
non_subscriber_counts = non_subscriber_counts.reindex(freq_options, fill_value=0)

fig = go.Figure()

fig.add_trace(go.Bar(
    x=freq_options,
    y=subscriber_counts.values,
    name='Подписчики'
))

fig.add_trace(go.Bar(
    x=freq_options,
    y=non_subscriber_counts.values,
    name='Неподписчики'
))

fig.update_layout(
    title="Распределение частоты покупок среди подписчиков и неподписчиков",
    xaxis_title="Частота покупок",
    yaxis_title="Количество покупателей",
    barmode='group',
    template="plotly"
)

fig.show()
```

```python
# Клиенты с подпиской чаще выбирают экспресс-доставку по сравнению с клиентами без подписки.
from scipy.stats import chi2_contingency

# Подсчитаем долю экспресс-доставки в каждой группе
express_delivery_subscribers = subscribers['Shipping Type'].value_counts(normalize=True).get('Express', 0)
express_delivery_non_subscribers = non_subscribers['Shipping Type'].value_counts(normalize=True).get('Express', 0)

print(f"Доля экспресс-доставки у подписчиков: {express_delivery_subscribers:.2%}")
print(f"Доля экспресс-доставки у неподписчиков: {express_delivery_non_subscribers:.2%}")

# Построим таблицу сопряженности для χ²-теста и проведем тест
contingency_table = pd.crosstab(data['Subscription Status'], data['Shipping Type'])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"χ²-статистика: {chi2:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

categories = ['Подписчики', 'Неподписчики']
values = [express_delivery_subscribers * 100, express_delivery_non_subscribers * 100]

fig = go.Figure(data=[
    go.Bar(name='Доля экспресс-доставки (%)', x=categories, y=values)
])

fig.update_layout(
    title="Сравнение доли экспресс-доставки между подписчиками и неподписчиками",
    xaxis_title="Группы покупателей",
    yaxis_title="Доля экспресс-доставки (%)",
    template="plotly"
)

fig.show()
```

```python
# Мужчины совершают покупки реже, чем женщины ('Frequency of Purchases').

male_buyers = data[data['Gender'] == 'Male']
female_buyers = data[data['Gender'] == 'Female']

# Вычисляем среднюю частоту покупок для мужчин и женщин
avg_frequency_male = male_buyers['Frequency Numeric'].mean()
avg_frequency_female = female_buyers['Frequency Numeric'].mean()

print(f"Средняя частота покупок у мужчин: {avg_frequency_male:.2f} раз в год")
print(f"Средняя частота покупок у женщин: {avg_frequency_female:.2f} раз в год")

t_stat, p_value = ttest_ind(
    male_buyers['Frequency Numeric'],
    female_buyers['Frequency Numeric'],
    equal_var=False  # Предполагаем разные дисперсии
)

print(f"T-статистика: {t_stat:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

# Шаг 5: Визуализация данных
categories = ['Мужчины', 'Женщины']
values = [avg_frequency_male, avg_frequency_female]

fig = go.Figure(data=[
    go.Bar(name='Средняя частота покупок (раз в год)', x=categories, y=values)
])

fig.update_layout(
    title="Сравнение средней частоты покупок между мужчинами и женщинами",
    xaxis_title="Группы покупателей",
    yaxis_title="Средняя частота покупок (раз в год)",
    template="plotly"
)

fig.show()


frequency_gender_counts = data.groupby(['Frequency of Purchases', 'Gender']).size().unstack(fill_value=0)

frequency_options = frequency_gender_counts.index.tolist()
male_counts = frequency_gender_counts['Male']
female_counts = frequency_gender_counts['Female']

fig = go.Figure()

fig.add_trace(go.Bar(
    x=frequency_options,
    y=male_counts,
    name='Мужчины'
))

fig.add_trace(go.Bar(
    x=frequency_options,
    y=female_counts,
    name='Женщины'
))

fig.update_layout(
    title="Распределение частоты покупок среди мужчин и женщин",
    xaxis_title="Частота покупок",
    yaxis_title="Количество покупателей",
    barmode='group',
    template="plotly"
)

fig.show()
```

```python
# Клиенты, использующие промокоды, чаще покупают товары со скидкой, чем без промокодов.

promo_users = data[data['Promo Code Used'] == 'Yes']
non_promo_users = data[data['Promo Code Used'] == 'No']

discount_with_promo = promo_users['Discount Applied'].value_counts(normalize=True).get('Yes', 0)
discount_without_promo = non_promo_users['Discount Applied'].value_counts(normalize=True).get('Yes', 0)

print(f"Доля товаров со скидкой при использовании промокодов: {discount_with_promo:.2%}")
print(f"Доля товаров со скидкой без использования промокодов: {discount_without_promo:.2%}")

contingency_table = pd.crosstab(data['Promo Code Used'], data['Discount Applied'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"χ²-статистика: {chi2:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

categories = ['С промокодом', 'Без промокода']
values = [discount_with_promo * 100, discount_without_promo * 100]

fig = go.Figure(data=[
    go.Bar(name='Доля товаров со скидкой (%)', x=categories, y=values)
])

fig.update_layout(
    title="Сравнение доли товаров со скидкой при использовании и без использования промокодов",
    xaxis_title="Группы покупателей",
    yaxis_title="Доля товаров со скидкой (%)",
    template="plotly"
)

fig.show()
```

```python
# Клиенты, имеющие более 20 предыдущих покупок, чаще совершают покупки (например, еженедельно), чем те, у кого меньше 20 покупок.
import plotly.express as px

more_than_20 = data[data['Previous Purchases'] > 20]
less_or_equal_20 = data[data['Previous Purchases'] <= 20]

avg_frequency_more_20 = more_than_20['Frequency Numeric'].mean()
avg_frequency_less_20 = less_or_equal_20['Frequency Numeric'].mean()

print(f"Средняя частота покупок у клиентов с более чем 20 покупками: {avg_frequency_more_20:.2f} раз в год")
print(f"Средняя частота покупок у клиентов с 20 или менее покупками: {avg_frequency_less_20:.2f} раз в год")

t_stat, p_value = ttest_ind(
    more_than_20['Frequency Numeric'],
    less_or_equal_20['Frequency Numeric'],
    equal_var=False  # Предполагаем разные дисперсии
)

print(f"T-статистика: {t_stat:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

data['Group'] = data['Previous Purchases'].apply(lambda x: 'Более 20 покупок' if x > 20 else '20 или менее покупок')

fig = px.box(
    data,
    x='Group',
    y='Frequency Numeric',
    points="all",
    title="Сравнение частоты покупок между группами",
    labels={"Frequency Numeric": "Частота покупок (раз в год)", "Group": "Группа"}
)

fig.update_layout(template="plotly")
fig.show()
```

```python
# Люди старше 40 лет чаще покупают товары из категории "Clothing", чем из других категорий.

data['Age Group'] = data['Age'].apply(lambda x: 'Older than 40' if x > 40 else '40 and Younger')

category_counts = data.groupby(['Age Group', 'Category']).size().unstack(fill_value=0)

clothing_older_40 = category_counts.loc['Older than 40', 'Clothing']
total_older_40 = category_counts.loc['Older than 40'].sum()
clothing_40_and_younger = category_counts.loc['40 and Younger', 'Clothing']
total_40_and_younger = category_counts.loc['40 and Younger'].sum()

clothing_share_older_40 = clothing_older_40 / total_older_40
clothing_share_40_and_younger = clothing_40_and_younger / total_40_and_younger

print(f"Доля покупок категории 'Clothing' у людей старше 40 лет: {clothing_share_older_40:.2%}")
print(f"Доля покупок категории 'Clothing' у людей 40 лет и младше: {clothing_share_40_and_younger:.2%}")

contingency_table = pd.crosstab(data['Age Group'], data['Category'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"χ²-статистика: {chi2:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

# Шаг 4: Визуализация данных (круговая диаграмма для группы старше 40 лет)
older_40_data = data[data['Age Group'] == 'Older than 40']
fig = px.pie(
    older_40_data,
    names='Category',
    title='Распределение категорий покупок у людей старше 40 лет',
    hole=0.4
)

fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
category_counts = data.groupby(['Age Group', 'Category']).size().unstack(fill_value=0)

category_proportions = category_counts.div(category_counts.sum(axis=1), axis=0)

fig = go.Figure()

for category in category_counts.columns:
    fig.add_trace(go.Bar(
        x=category_proportions.index,
        y=category_proportions[category],
        name=category
    ))

fig.update_layout(
    title="Распределение категорий покупок по возрастным группам",
    xaxis_title="Возрастная группа",
    yaxis_title="Доля покупок",
    barmode='stack',  # Измените на 'group' для группированного графика
    template="plotly"
)

fig.show()
```

```python
# Покупатели, оплачивающие покупки наличными, реже используют скидки, чем покупатели, 
# использующие электронные платежи.

cash_users = data[data['Payment Method'] == 'Cash']
electronic_users = data[data['Payment Method'] != 'Cash']  # Все остальные способы оплаты

discount_cash = cash_users['Discount Applied'].value_counts(normalize=True).get('Yes', 0)
discount_electronic = electronic_users['Discount Applied'].value_counts(normalize=True).get('Yes', 0)

print(f"Доля покупок со скидкой для оплаты наличными: {discount_cash:.2%}")
print(f"Доля покупок со скидкой для электронных платежей: {discount_electronic:.2%}")

contingency_table = pd.crosstab(data['Payment Method'] == 'Cash', data['Discount Applied'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"χ²-статистика: {chi2:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

categories = ['Наличные', 'Электронные платежи']
values = [discount_cash * 100, discount_electronic * 100]

fig = go.Figure(data=[
    go.Bar(name='Доля покупок со скидкой (%)', x=categories, y=values)
])

fig.update_layout(
    title="Сравнение доли покупок со скидкой между методами оплаты",
    xaxis_title="Метод оплаты",
    yaxis_title="Доля покупок со скидкой (%)",
    template="plotly"
)

fig.show()
```

```python
# Клиенты с подпиской оставляют более высокие отзывы (по рейтингу), чем клиенты без подписки.

avg_rating_subscribers = subscribers['Review Rating'].mean()
avg_rating_non_subscribers = non_subscribers['Review Rating'].mean()

print(f"Средний рейтинг подписчиков: {avg_rating_subscribers:.2f}")
print(f"Средний рейтинг неподписчиков: {avg_rating_non_subscribers:.2f}")

# Шаг 3: Провести t-тест для проверки значимости различий
t_stat, p_value = ttest_ind(
    subscribers['Review Rating'],
    non_subscribers['Review Rating'],
    equal_var=False  # Предполагаем разные дисперсии
)

print(f"T-статистика: {t_stat:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

# Шаг 4: Визуализация данных (boxplot)
data['Group'] = data['Subscription Status'].apply(lambda x: 'Подписчики' if x == 'Yes' else 'Неподписчики')

fig = px.box(
    data,
    x='Group',
    y='Review Rating',
    points="all",
    title="Сравнение рейтингов покупок между подписчиками и неподписчиками",
    labels={"Review Rating": "Рейтинг покупки", "Group": "Группа"}
)

fig.update_layout(template="plotly")
fig.show()
```

```python
# Товары серого цвета чаще покупаются зимой, чем в другие сезоны.

gray_items = data[data['Color'] == 'Gray']

gray_counts_by_season = gray_items['Season'].value_counts()
total_counts_by_season = data['Season'].value_counts()

gray_proportions = gray_counts_by_season / total_counts_by_season

print("Доля покупок серого цвета по сезонам:")
print(gray_proportions)

contingency_table = pd.crosstab(data['Color'] == 'Gray', data['Season'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"χ²-статистика: {chi2:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

# Шаг 5: Визуализация данных (столбчатый график)
seasons = gray_proportions.index
values = gray_proportions.values * 100  # Процентное представление

fig = go.Figure(data=[
    go.Bar(name='Доля покупок серого цвета (%)', x=seasons, y=values)
])

fig.update_layout(
    title="Доля покупок серого цвета по сезонам",
    xaxis_title="Сезон",
    yaxis_title="Доля покупок серого цвета (%)",
    template="plotly"
)

fig.show()
```

```python
# Покупки с бесплатной доставкой имеют в среднем более низкую стоимость, чем покупки с экспресс-доставкой.

free_shipping = data[data['Shipping Type'] == 'Free Shipping']
express_shipping = data[data['Shipping Type'] == 'Express']

avg_price_free = free_shipping['Purchase Amount (USD)'].mean()
avg_price_express = express_shipping['Purchase Amount (USD)'].mean()

print(f"Средняя стоимость покупок с бесплатной доставкой: {avg_price_free:.2f} USD")
print(f"Средняя стоимость покупок с экспресс-доставкой: {avg_price_express:.2f} USD")

t_stat, p_value = ttest_ind(
    free_shipping['Purchase Amount (USD)'],
    express_shipping['Purchase Amount (USD)'],
    equal_var=False  # Предполагаем разные дисперсии
)

print(f"T-статистика: {t_stat:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

data['Shipping Group'] = data['Shipping Type'].apply(
    lambda x: 'Бесплатная доставка' if x == 'Free Shipping' else ('Экспресс-доставка' if x == 'Express' else 'Другие')
)

fig = px.box(
    data[data['Shipping Group'].isin(['Бесплатная доставка', 'Экспресс-доставка'])],
    x='Shipping Group',
    y='Purchase Amount (USD)',
    points="all",
    title="Сравнение стоимости покупок между типами доставки",
    labels={"Purchase Amount (USD)": "Стоимость покупки (USD)", "Shipping Group": "Тип доставки"}
)

fig.update_layout(template="plotly")
fig.show()
```

```python
# Средняя стоимость покупок, сделанных летом, ниже, чем зимой.

filtered_data = data[data['Season'].isin(['Summer', 'Winter'])]

summer_data = filtered_data[filtered_data['Season'] == 'Summer']
winter_data = filtered_data[filtered_data['Season'] == 'Winter']

avg_price_summer = summer_data['Purchase Amount (USD)'].mean()
avg_price_winter = winter_data['Purchase Amount (USD)'].mean()

print(f"Средняя стоимость покупок летом: {avg_price_summer:.2f} USD")
print(f"Средняя стоимость покупок зимой: {avg_price_winter:.2f} USD")

t_stat, p_value = ttest_ind(
    summer_data['Purchase Amount (USD)'],
    winter_data['Purchase Amount (USD)'],
    equal_var=False  # Предполагаем разные дисперсии
)

print(f"T-статистика: {t_stat:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

fig = px.box(
    filtered_data,
    x='Season',
    y='Purchase Amount (USD)',
    points="all",
    title="Сравнение стоимости покупок между летом и зимой",
    labels={"Purchase Amount (USD)": "Стоимость покупки (USD)", "Season": "Сезон"}
)

fig.update_layout(template="plotly")
fig.show()
```

```python
# В зимний сезон покупатели чаще выбирают товары "Jackets".

winter_data = data[data['Season'] == 'Winter']
other_seasons_data = data[data['Season'] != 'Winter']

winter_jackets_count = winter_data['Item Purchased'].value_counts().get('Jacket', 0)
winter_total_count = winter_data['Item Purchased'].count()

other_seasons_jackets_count = other_seasons_data['Item Purchased'].value_counts().get('Jacket', 0)
other_seasons_total_count = other_seasons_data['Item Purchased'].count()

winter_jackets_proportion = winter_jackets_count / winter_total_count
other_seasons_jackets_proportion = other_seasons_jackets_count / other_seasons_total_count

print(f"Доля товаров 'Jacket' в зимний сезон: {winter_jackets_proportion:.2%}")
print(f"Доля товаров 'Jacket' в другие сезоны: {other_seasons_jackets_proportion:.2%}")

contingency_table = pd.crosstab(data['Season'] == 'Winter', data['Item Purchased'] == 'Jacket')

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"χ²-статистика: {chi2:.2f}")
print(f"P-значение: {p_value:.4f}")

interpret(p_value)

categories = ['Winter', 'Other Seasons']
values = [winter_jackets_proportion * 100, other_seasons_jackets_proportion * 100]

fig = go.Figure(data=[
    go.Bar(name='Доля товаров "Jacket" (%)', x=categories, y=values)
])

fig.update_layout(
    title="Доля товаров 'Jacket' в зимний сезон и в другие сезоны",
    xaxis_title="Сезоны",
    yaxis_title="Доля товаров 'Jacket' (%)",
    template="plotly"
)

fig.show()

fig_pie = px.pie(
    winter_data,
    names='Item Purchased',
    title='Распределение категорий покупок в зимний сезон',
    hole=0.4
)

fig_pie.update_traces(textinfo='percent+label')
fig_pie.show()
```

```python
# посчитайте наш сезонный mau (уникальных пользователей за сезон) и его динамику.
# Считала что Fall - первый сезон 

# Сезонный MAU (уникальные пользователи за сезон)
seasonal_mau = data.groupby('Season')['Customer ID'].nunique().reset_index()
seasonal_mau.rename(columns={'Customer ID': 'MAU'}, inplace=True)

# Упорядочение сезонов: Fall → Winter → Spring → Summer
season_order = ['Fall', 'Winter', 'Spring', 'Summer']
seasonal_mau['Season'] = pd.Categorical(seasonal_mau['Season'], categories=season_order, ordered=True)
seasonal_mau = seasonal_mau.sort_values('Season')

# Добавить динамику MAU (процентное изменение по сравнению с предыдущим сезоном)
seasonal_mau['MAU Change (%)'] = seasonal_mau['MAU'].pct_change() * 100

print("Сезонный MAU с динамикой:")
print(seasonal_mau)

# Построить круговой график для сезонного MAU
fig_pie = px.pie(
    seasonal_mau,
    names='Season',
    values='MAU',
    title="Распределение MAU (уникальные пользователи) по сезонам",
    hole=0.4  # Добавляем отверстие для "donut" графика
)

fig_pie.update_traces(
    textinfo='percent+label',  # Отображать процент и название сезона
    pull=[0.1 if mau == seasonal_mau['MAU'].max() else 0 for mau in seasonal_mau['MAU']]  # Выделить сектор с максимальным MAU
)

fig_pie.update_layout(template="plotly")
fig_pie.show()

# Построить график динамики MAU
fig_change = px.area(
    seasonal_mau,
    x='Season',
    y='MAU Change (%)',
    title="Динамика изменения MAU по сезонам",
    labels={"Season": "Сезон", "MAU Change (%)": "Изменение MAU (%)"},
    text='MAU Change (%)'
)

fig_change.update_traces(
    texttemplate='%{y:.2f}%', 
    textposition="top center", 
    line=dict(shape='linear', width=2),
    fill='tozeroy'
)

fig_change.update_layout(template="plotly")
fig_change.show()
```

### <h1>Задание 3</h1> 

У нас есть полный каталог всех случаев, когда кто-то проклинал или истекал кровью в фильме Квентина Тарантино. https://github.com/fivethirtyeight/data/tree/master/tarantino

Определите:

* число проклятий, летальных исходов и их соотношение по фильмам 

* частоту употребления конкретных проклятий 

* распределение времени между проклятиями 

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Загрузка данных
url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/tarantino/tarantino.csv"
data = pd.read_csv(url)

data
```

```python
data['movie'].unique()
```

```python
# Подсчитать проклятия и смерти по фильмам
curse_death_summary = data.groupby('movie').agg({
    'type': [
        lambda x: (x == 'word').sum(),  # Число проклятий
        lambda x: (x == 'death').sum()  # Число смертей
    ]
}).reset_index()
curse_death_summary.columns = ['movie', 'curse_count', 'death_count']

# Объединить проклятия и смерти в один столбец
curse_death_summary['count_summary'] = curse_death_summary.apply(
    lambda row: f"Проклятия: {row['curse_count']}, Смерти: {row['death_count']}", axis=1
)

print("Объединенные данные:")
print(curse_death_summary[['movie', 'count_summary']])

# Визуализация проклятий и смертей
fig = px.bar(
    curse_death_summary,
    x='movie',
    y=['curse_count', 'death_count'],
    title="Проклятия и смерти по фильмам",
    labels={"value": "Количество", "variable": "Тип", "movie": "Фильм"},
    barmode='stack',
    template="plotly"
)

fig.update_layout(
    xaxis_title="Фильм",
    yaxis_title="Количество",
    legend_title="Тип",
    showlegend=True
)
fig.show()
```

```python
# 2. Частота употребления конкретных проклятий
curse_frequency = data[data['type'] == 'word']['word'].value_counts().reset_index()
curse_frequency.columns = ['curse', 'count']

print("Частота употребления конкретных проклятий:")
print(curse_frequency)

# Визуализация частоты употребления проклятий
fig_curse_freq = px.bar(
    curse_frequency.head(20),
    x='curse',
    y='count',
    title="Топ-10 проклятий по частоте",
    labels={"curse": "Проклятие", "count": "Частота"},
    template="plotly"
)
fig_curse_freq.show()
```

```python
# Шаг 1: Фильтрация только проклятий
curse_data = data[data['type'] == 'word']

# Шаг 2: Сортировка по времени и расчет времени между проклятиями для каждого фильма
def calculate_time_between_curses(df):
    df = df.sort_values(by='minutes_in')
    df['time_between_curses'] = df['minutes_in'].diff()
    return df

curse_data = curse_data.groupby('movie').apply(calculate_time_between_curses).reset_index(drop=True)

# Удаляем строки с NaN (первая строка каждого фильма)
curse_data = curse_data.dropna(subset=['time_between_curses'])

# Просмотр данных
print("Распределение времени между проклятиями по фильмам:")
print(curse_data[['movie', 'time_between_curses']].groupby('movie').describe())

# Шаг 3: Построение графиков для каждого фильма
movies = curse_data['movie'].unique()

for movie in movies:
    movie_data = curse_data[curse_data['movie'] == movie]
    fig = px.histogram(
        movie_data['time_between_curses'],
        nbins=15,
        title=f"Распределение времени между проклятиями — {movie}",
        labels={"value": "Время между проклятиями (минуты)", "count": "Количество"},
        template="plotly"
    )
    fig.update_layout(
        xaxis_title="Время между проклятиями (минуты)",
        yaxis_title="Частота",
        bargap=0.2
    )
    fig.show()
```

```python
# Построение общего распределения времени между проклятиями
all_curse_times = curse_data['time_between_curses']

fig_overall = px.histogram(
    all_curse_times,
    nbins=30,
    title="Общее распределение времени между проклятиями",
    labels={"value": "Время между проклятиями (минуты)", "count": "Количество"},
    template="plotly"
)

fig_overall.update_layout(
    xaxis_title="Время между проклятиями (минуты)",
    yaxis_title="Частота",
    bargap=0.2
)
fig_overall.show()
```

```python
fig_violin = px.violin(
    all_curse_times,
    box=True,  # Добавляет box plot внутри
    points="all",  # Показывает все точки
    title="Violin Plot: Время между проклятиями",
    labels={"value": "Время между проклятиями (минуты)"},
    template="plotly"
)
fig_violin.update_layout(
    yaxis_title="Время между проклятиями (минуты)"
)
fig_violin.show()
```
